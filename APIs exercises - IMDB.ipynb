```python
# List of word to searching
keywords=['Story', 'Horror', 'Home', 'Start', 'Country', 'Man', 'Woman', 'Life', 
          'World', 'Night', 'Game', 'Love', 'Sky', 'Happy', 'Family', 'Christmas']
```

We iterate over the results. We store only the results with type 'movie'.

```python
imdb_ids = []

for keyword in keywords:
    url ="http://www.omdbapi.com/?s={0}".format(keyword)
    content = requests.get(url).json()
    search = content['Search']
    for s in search:
        if(s['Type'] =='movie'):
            imdb_ids.append(s['imdbID'])
print(imdb_ids)
```

Define the csv file. The headers will be 'Genre', 'Description', 'Type', 'Title' and 'imdbRating'.

```python
csv_imdb = []
csv_hearders = ['Title', 'Genre', 'Description', 'Type', 'imdbRating']
csv_imdb.append(csv_hearders)
```

For each movie, we get its info from the API to store in the csv file.

```python
write_file = open('imdb.csv', 'w')
writer = csv.writer(write_file, delimiter=',')
 
for id in imdb_ids:
    url = "http://www.omdbapi.com/?i={0}".format(id)
    content = requests.get(url).json()
    genre = content['Genre']
    description = content['Plot']
    type = content['Type']
    title = content['Title']
    imdb_rating = content['imdbRating']
    csv_imdb.append([title, genre, description, type, imdb_rating])
    

# Save the info in the csv file
for row in csv_imdb:
    try:
        writer.writerow(row)
    except:
        pass

# Close the csv file
write_file.close()
```
