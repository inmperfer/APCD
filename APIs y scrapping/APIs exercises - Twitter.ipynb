```python
import csv
import twitter
import re
```

```python
# Twitter credentials
CONSUMER_KEY = ''
CONSUMER_SECRET = ''
OAUTH_TOKEN = ''
OAUTH_TOKEN_SECRET = ''
```

Set up the auth for the twitter package.

```python
auth = twitter.oauth.OAuth(OAUTH_TOKEN, OAUTH_TOKEN_SECRET,CONSUMER_KEY, CONSUMER_SECRET)
twitter_api = twitter.Twitter(auth=auth)
```

Execute the tweets search and print the keys of the results.

```python
search_word = '#navidad2016'
count = 500 # Max number of results

search_results = twitter_api.search.tweets(q=search_word, count=count)
print(search_results.keys())
statuses = search_results['statuses']
```

Define the csv file and the headers to store. By default, you will save user, text and hashtags.

```python
import csv

csv_tweets = []
csv_headers=["user", "text", "hashtags", "retweet_count", "user_mentions"]
csv_tweets.append(csv_headers)
```

For each tweet, replace the user for the word USER and the retweets for the word RETWEET. Also, get the hashtags. Finally, save the info in the csv file.

```python
print(statuses[0].keys())
```

```python
hashtags=[]

write_file = open('twitter.csv', 'w')
writer = csv.writer(write_file, delimiter=',')

for elem in statuses:
    # Get the tweet text
    text = elem['text']
    # Get the user
    user = elem['user']['screen_name']
    # Get the hashtags
    hashtags = re.findall(r'#[\w.-]+', text)
    # Replace retweets for the word RETWEET
    tweet = re.sub(r'RT @[\w.-]+', r'RETWEET', text)
    # Replace the user for the word USER
    tweet = re.sub(r'@[\w.-]+', r'USER', tweet)
    retweet_count = elem['retweet_count']
    user_mentions = len(elem['entities']['user_mentions'])
     
    csv_tweets.append([user, tweet, hashtags, retweet_count, user_mentions])
      
# Save the info in the csv file
for row in csv_tweets:
    try:
        writer.writerow(row)
    except:
        pass

# Close the csv file
write_file.close()
```
